### 机械学习诊断法（？）
- **判断是偏差还是方差** 算法运行不理想时，可能是方差（过拟合）过大或者是偏差（欠拟合）过大
- **判断依据**训练误差随着多项式次数的增大而减小，而验证误差则像一个二次函数
- **判断方法**当训练误差与验证误差都大时，就是偏差大（欠拟合），训练小验证大就是方差大（过拟合）
#### 正则化和偏差方差
- 正则化参数大，会出现高偏差，正则化参数过小会出现过拟合高方差的情况，所以正则化系数对算法影响很大
- **正则化系数的选择**，从0开始以步长依次翻倍增加（第一次为0.01），将对应的系数得到一系列参数向量 theta_n 用交叉验证集评价，选取选取误差最小的最合适**看是否合适的时候代价函数中不包括正则化，方便找问题但是找theta时有正则化**
- **正则化与训练误差验证误差的关系**lambda（正则化参数）过大至超过一个阈值时，因为惩罚项太大，导致丢失太多特征，会出现欠拟合现象

```python

from matplotlib import pyplot as plt
import matplotlib
import numpy as np
# 定义正则化参数
lambda_values = np.linspace(0.01,10,500)
# 定义训练误差与验证误差的函数
def J_train(lambda_val):
    return 0.1*lambda_val**2+0.01
def J_cv(lambda_val):
    return (lambda_val-5) ** 2 + 4
# 计算误差
J_train_values = [J_train(l) for l in lambda_values]
J_cv_values = [J_cv(i) for i in lambda_values]
# 创建图形
plt.figure(figsize=(8,6))
# 绘制误差曲线
plt.plot(lambda_values, J_train_values, label=r'$J_{\text{train}}(\theta)$', color='blue')
plt.plot(lambda_values, J_cv_values, label=r'$J_{\text{cv}}(\theta)$', color='magenta')
# 添加图例
plt.legend()
# 添加标题与标签
plt.title(' bewteen lambda and J_error ')
plt.xlabel(r'$\lambda$')
plt.ylabel('Error')
# 显示图形
plt.grid(True)
plt.show()
```
