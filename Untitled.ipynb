## numpy的注意点

-  a=b 完全不复制，a和b相互影响,可变对象包括列表（  list  ）、字典（  dict  ）、集合（  set  ）等,不可变对象包括数字（如int、float）、字符串（str）、元组（tuple ）等。

- a = b[:],视图的操作，一种切片，会创建新的对象a，但是a的数据完全由b保管，他们两个的数据变化是一致的，

- a = b.copy(),复制，a和b互不影响
